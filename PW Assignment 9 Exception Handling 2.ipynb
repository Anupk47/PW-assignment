Q1. Explain why we have to use the Exception class while creating a Custom Exception.

```python
#In Python, all built-in exceptions are derived from the BaseException class, and more commonly from its subclass Exception. 
#When creating custom exceptions, we inherit from the Exception class to ensure our custom exception integrates with Pythonâ€™s exception-handling framework
#example
class CustomeERROR(Exception):
    pass
try:
    raise CustomeERROR("Something Wrong")
except CustomeERROR as e:
    print(e)
```
